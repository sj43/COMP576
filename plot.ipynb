```python
import matplotlib
import matplotlib.pyplot as plt
plt.plot([1,2,3,4], [1,2,7,14])
plt.axis([0, 6, 0, 20])
plt.show()
```

```python
import numpy as np

matplotlib.use('Agg')


fig = plt.figure(figsize=(5, 5), dpi=100)
axes = fig.add_axes([0.01, 0.01, .50, 0.50])
X = np.linspace(0, 3, 1000)
Y = np.sin(5*np.pi*X)
plt.plot(X, Y, lw=2)
plt.ylim(-1.2, 1.2)
plt.grid()

plt.show()
```
